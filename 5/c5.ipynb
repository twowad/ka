```python
import pandas as pd
import swifter
```

```python
df=pd.read_csv('forestfires.csv')
df
```

```python
def map_function(row):
    result=row['X'] * row['wind']
    return result
```

```python
df['map_result']=df.swifter.apply(map_function,axis=1)
df
```

```python
reduce_result = df['map_result'].sum()
reduce_result
```
